## Introduction to Statistics

```python
# datasets
DA = np.array([23, 24, 24, 24, 25, 25, 25, 25, 26, 26, 26, 27])
DB = np.array([2, 5, 10, 12, 15, 20, 30, 35, 38, 40, 45, 48])
```

```python
# numero de elementos
n_A = len(DA)
n_B = len(DB)

# suma de elementos
s_A = np.sum(DA)
s_B = np.sum(DB)

print(f'Dataset A: n(A) = {n_A}, s(A) = {s_A}')
print(f'Dataset B: n(B) = {n_B}, s(B) = {s_B}')
```

## Tendencia central

```python
# media
A_media = s_A / n_A
B_media = s_B / n_B

print(f'Dataset A: media = {A_media}')
print(f'Dataset B: media = {B_media}')
```

```python
# mediana
A_mediana = np.median(DA)
B_mediana = np.median(DB)

print(f'Dataset A: mediana = {A_mediana}')
print(f'Dataset B: mediana = {B_mediana}')
```

```python
# moda
A_counts = np.bincount(DA)
A_moda = np.argmax(A_counts)

print(f'Dataset A: moda = {A_moda}')
```

## Dispersión

```python
# Histograma dataset A
bins = np.arange(min(DA), max(DA) + 2) - 0.5

plt.figure(figsize=(4, 4))
plt.hist(DA, bins=bins, color="skyblue", edgecolor='k', rwidth=0.4)
plt.hist(DA, bins=bins, edgecolor='k', histtype='step', alpha=0.2)
plt.title('Dataset A')
plt.xlabel('Llamaradas solares')
plt.ylabel('Conteo')
plt.ylim(0, 5)
plt.savefig('DA_histograma.png', bbox_inches='tight')
```

```python
# Histograma dataset B
from collections import Counter
counts = Counter(DB)

plt.figure(figsize=(4, 4))
plt.bar(counts.keys(), counts.values(), width=1.5, color='#F08080', edgecolor='k')
plt.title('Dataset B')
plt.xlabel('Llamaradas solares')
plt.ylabel('Conteo')
plt.ylim(0, 5)
plt.savefig('DB_histograma.png', bbox_inches='tight')
```

## Hubble

```python
data_Hubble = np.array([
    [11, 20, 28, 36, 45, 54, 66, 73, 87, 96, 110, 122],
    [1050, 1400, 1850, 2750, 3050, 3950, 4550, 5450, 5900, 7050, 7600, 8350]
])
```

```python
print(data_Hubble.T)
```

```python
x = data_Hubble[0]
y = data_Hubble[1]
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x, y, 'ko', ms=3)
plt.xlabel('Distancia (Mpc)')
plt.ylabel(r'Velocidad (km s$^{-1}$)')

plt.plot([x[0], x[-1]], [y[0], y[-1]], 'b', lw=0.5)

plt.savefig('linear_reg-2.png', bbox_inches='tight')
```

```python
print((x*y))
```

```python
np.sum(y) / 12
```

```python
np.sum(x**2)
```

```python
np.sum(x*y)
```

```python
m = (np.sum(x * y) - 12 * np.mean(x) * np.mean(y)) / (np.sum(x**2) - 12 * np.mean(x)**2)
print(m)
```

```python
b = np.mean(y) - m * np.mean(x)
print(b)
```

## Kepler

```python
data_kepler = np.array([
    # r (AU),   T (years)
    [0.41, 0.27],
    [0.70, 0.60],
    [0.98, 0.99],
    [1.55, 1.95],
    [2.30, 3.55],
    [5.25, 11.70],
    [9.60, 29.50],
    [12.8, 45.00],
    [15.5, 61.00],
    [19.0, 82.00],
    [23.5, 113.00],
    [27.0, 141.00],
    [30.5, 168.00]
])
```

```python
a = data_kepler[:, 0]
T = data_kepler[:, 1]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(a, T, 'ko', ms=3)
plt.xlabel('a (AU)')
plt.ylabel(r'T (años)')

#plt.plot([x[0], x[-1]], [y[0], y[-1]], 'b', lw=0.5)
plt.grid(alpha=0.5)
plt.xlim(-5, 35)
plt.ylim(-25, 225)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.plot(a**3, T**2, 'ko', ms=3)
plt.xlabel(r'a$^3$ (AU$^3$)')
plt.ylabel(r'T$^2$ (años$^2$)')

#plt.plot([x[0], x[-1]], [y[0], y[-1]], 'b', lw=0.5)
plt.grid(alpha=0.5)
plt.show()
```
