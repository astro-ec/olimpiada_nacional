```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import scienceplots
plt.style.use(['science', 'nature', 'std-colors'])

import matplotlib.pylab as pylab
params = {'legend.fontsize': 12,
         'axes.labelsize': 12,
         'xtick.labelsize': 12,
         'ytick.labelsize': 12}
pylab.rcParams.update(params)
plt.rcParams.update({'font.size': 12})
```

```python
# datasets
DA = np.array([23, 24, 24, 24, 25, 25, 25, 25, 26, 26, 26, 27])
DB = np.array([2, 5, 10, 12, 15, 20, 30, 35, 38, 40, 45, 48])
```

```python
# numero de elementos
n_A = len(DA)
n_B = len(DB)

# suma de elementos
s_A = np.sum(DA)
s_B = np.sum(DB)

print(f'Dataset A: n(A) = {n_A}, s(A) = {s_A}')
print(f'Dataset B: n(B) = {n_B}, s(B) = {s_B}')
```

## Tendencia central

```python
# media
A_media = s_A / n_A
B_media = s_B / n_B

print(f'Dataset A: media = {A_media}')
print(f'Dataset B: media = {B_media}')
```

```python
# mediana
A_mediana = np.median(DA)
B_mediana = np.median(DB)

print(f'Dataset A: mediana = {A_mediana}')
print(f'Dataset B: mediana = {B_mediana}')
```

```python
# moda
A_counts = np.bincount(DA)
A_moda = np.argmax(A_counts)

print(f'Dataset A: moda = {A_moda}')
```

## Dispersión

```python
# Histograma dataset A
bins = np.arange(min(DA), max(DA) + 2) - 0.5

plt.figure(figsize=(4, 4))
plt.hist(DA, bins=bins, color="skyblue", edgecolor='k', rwidth=0.4)
plt.hist(DA, bins=bins, edgecolor='k', histtype='step', alpha=0.2)
plt.title('Dataset A')
plt.xlabel('Llamaradas solares')
plt.ylabel('Conteo')
plt.ylim(0, 5)
plt.savefig('DA_histograma.png', bbox_inches='tight')
```

```python
# Histograma dataset B
from collections import Counter
counts = Counter(DB)

plt.figure(figsize=(4, 4))
plt.bar(counts.keys(), counts.values(), width=1.5, color='#F08080', edgecolor='k')
plt.title('Dataset B')
plt.xlabel('Llamaradas solares')
plt.ylabel('Conteo')
plt.ylim(0, 5)
plt.savefig('DB_histograma.png', bbox_inches='tight')
```
